```python
import pandas as pd
```

```python
data ={
    'Yil' : [2021,2020,2019,2018,2017,2016,2015,2010],
    'Aholi soni' :[33.9, 33.5, 33.6, 33.2, 35.6, 33.1, 33.7, 33],
    'Temp':[1.54, 1.26, 1.32, 1.27, 1.96, 1.65, 1.56, 1.42],
    'Zichlik':[76.3, 75.5, 78.1, 74.3, 77.7, 78.5, 79.6, 71.2]
      }
df = pd.DataFrame(data)
df
```

```python
df.index = [21,20,19,18,17,16,15,10]
df
```

```python
# Reindex
ind =[21,20,19,18,17,16,15,10]
ind.sort()
ind
```

```python
df = df.reindex(ind)
df
```

```python
df.reindex([10,15,20])
```

```python
df.reindex(index = range(10,22))
```

```python
df.reindex(index = range(10,22) , method='nearest')
```
